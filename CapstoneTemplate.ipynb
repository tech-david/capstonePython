```python
## Get access to s3 bucket

import pandas as pd
import boto3

ACCESS=pd.read_csv("/content/drive/MyDrive/Capstone/Control/jupyters3access.csv")

AWS_S3_BUCKET="capstoneds2022data"
AWS_ACCESS_KEY_ID=ACCESS['Access key ID'].values[0]
AWS_SECRET_ACCESS_KEY=ACCESS['Secret access key'].values[0]


s3=boto3.resource('s3',
                  aws_access_key_id=AWS_ACCESS_KEY_ID,
                  aws_secret_access_key=AWS_SECRET_ACCESS_KEY)
s3.Bucket(AWS_S3_BUCKET).download_file('Data Sources/EIA/Table_9.10_Natural_Gas_Prices.xlsx', 
                                       '/content/drive/MyDrive/Capstone/tmp/natural_gas')

```
